```python
import datetime
import d22a
import matplotlib.pyplot as plt
```

```python
# Get start datetime
start_dt = datetime.datetime.now()
```

```python
# Watermark
print(d22a.get_watermark())
```

```python
# Modify plot settings
plt.rcParams['axes.titlesize'] = 'xx-large'
plt.rcParams['axes.labelsize'] = 'xx-large'
plt.rcParams['xtick.labelsize'] = 'large'
plt.rcParams['ytick.labelsize'] = 'large'
plt.rcParams['legend.fontsize'] = 'large'
```

```python
# Apply MCDC to Z, for use in presentation
fig = d22a.composite_compare_methods_timeseries(variable='Z', degrees=('agnostic', 'linear'))
axs = fig.get_axes()
axs[0].set_title('Drift samples')
axs[1].set_title('Drift-corrected control')
axs[2].set_title('Drift-corrected historical')
for ax in axs:
    leg = ax.legend(fontsize='large')
    leg = d22a.legend_min_alpha_linewidth(leg)
_ = d22a.name_save_fig(fig, fso='o')
```

```python
# Apply MCDC to Z for a different ESM, for use in presentation
fig = d22a.composite_compare_methods_timeseries(esm='CMCC-ESM2_r1i1p1f1', variable='Z', degrees=('agnostic', 'linear'))
axs = fig.get_axes()
axs[0].set_title('Drift samples')
axs[1].set_title('Drift-corrected control')
axs[2].set_title('Drift-corrected historical')
for ax in axs:
    leg = ax.legend(fontsize='large')
    leg = d22a.legend_min_alpha_linewidth(leg)
_ = d22a.name_save_fig(fig, fso='o')
```

```python
# Get end datetime
end_dt = datetime.datetime.now()
# Calculate run timedelta
run_td = end_dt - start_dt
# Print timing information
print(f"Start:     {start_dt.strftime('%Y-%m-%d %H:%M:%S')}")
print(f"End:       {end_dt.strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Wall time: {run_td.seconds} s")
```
